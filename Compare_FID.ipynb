```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DCGAN = np.load('Results/FID_score/DC_FID.npy')
WGANGP = np.load('Results/FID_score/WGANGP_FID.npy')
ACGAN = np.load('Results/FID_score/ACGAN_FID.npy')
```

```python
plt.figure(figsize=(12, 6))

# Plot FID scores with different styles, markers, and colors
plt.plot(DCGAN, label="DCGAN", color='blue', linestyle='-', marker='o', markersize=4)
plt.plot(WGANGP, label="WGAN", color='green', linestyle='--', marker='s', markersize=4)
plt.plot(ACGAN, label="ACGAN", color='red', linestyle='-.', marker='^', markersize=4)

# Title and labels
plt.title("FID Scores Comparison for DCGAN vs WGAN vs ACGAN", fontsize=18)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("FID", fontsize=14)

# Legend with fontsize
plt.legend(loc='upper right', fontsize=12)

# Grid with a lighter shade
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Tight layout
plt.tight_layout()

# Save and show the plot
plt.savefig('Results/FID_Compare.jpg', format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```
